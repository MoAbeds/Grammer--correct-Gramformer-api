```python
from gramformer import Gramformer
from fastapi import FastAPI
from typing import Optional
```

```python
gf = Gramformer(models = 1, use_gpu=False) # 1=corrector, 2=detector
```

```python
gf.correct('My camera batery a dead')
```

```python
sentences = [
    'helle my friend', 
    'World is flat', 
    'Red a color', 
    'I wish my Computer was run faster.'
]
```

```python
for false_sentence in sentences:
  correct_sentences = gf.correct(false_sentence,max_candidates=1)
  print('[input]', false_sentence)
  for correct_sentence in correct_sentences:
    print('[correct]',correct_sentence)
    print("-" *100)

```
